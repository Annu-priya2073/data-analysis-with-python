Interpretation

# PCA (Dimensionality Reduction)

PCA (Principal Component Analysis)

PCA (Principal Component Analysis) is a **dimensionality reduction** technique used to reduce the number of features in a dataset while retaining most of the important information (variance).

It transforms the original numerical features into a new set of variables called **Principal Components (PCs)**:
- **PC1** captures the maximum variance
- **PC2** captures the second maximum variance
- and so on...

```python
# Apply PCA on Standardized Data
from sklearn.decomposition import PCA

pca = PCA()
X_pca = pca.fit_transform(X_standardized)

# explained variance ratio
explained_variance = pca.explained_variance_ratio_
explained_variance
```

# Plot explained variance

```python
plt.figure(figsize=(8, 4))
plt.plot(np.cumsum(explained_variance), marker='o')
plt.title("Cumulative Explained Variance by PCA Components")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.grid(True)
plt.savefig("images/pca_explained_variance.png", dpi=300, bbox_inches="tight")
plt.show()
```

Interpretation
- The 1st PCA component explains about 36% of the total variance, meaning a large amount of information is captured in the first component itself.

- The curve increases steadily, showing that each additional component adds more information.

- Around 10 components, the model captures approximately 83% of the total variance.

- Around 13 components, the variance explained reaches about 90%, which is usually considered enough for dimensionality reduction.
  After that, the curve becomes almost flat, meaning adding more components gives only small improvement.
  
  So, the dataset can be reduced to around 10–13 PCA components while still preserving most of the important information.

```python
# Choose PCA components ( 2 components)
pca_2 = PCA(n_components=2)
X_pca_2 = pca_2.fit_transform(X_standardized)

X_pca_2[:5]
```

Interpretation
- PCA reduces the number of features while keeping important information.
This helps in clustering and visualization.

# K-Means Clustering (Customer Segmentation)

K-Means is an **unsupervised machine learning algorithm** used for clustering.  
It groups similar data points into **K clusters** based on their feature similarity.

In this project, K-Means clustering is applied to perform **customer segmentation**, where customers are grouped according to their engagement and purchasing behavior.


Why K-Means Clustering?
- Helps identify different customer groups/segments
- Useful for targeted marketing and business strategies
- Groups customers with similar behavior patterns

```python
from sklearn.cluster import KMeans

inertia = []
k_range = range(1, 11)

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_pca_2)
    inertia.append(km.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(k_range, inertia, marker="o")
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.grid(True)
plt.savefig("images/kmeans_clusters.png", dpi=300, bbox_inches="tight")
plt.show()
```

Interpretation of your Elbow Method graph (Optimal K)

- The graph shows Inertia decreases sharply from K=1 to K=3, meaning clustering improves a lot when we increase clusters up to 3.

- After K=3, the decrease in inertia becomes slow and gradual, which means adding more clusters does not improve the clustering much.

- The point where the curve changes from steep to slow is called the Elbow point.

- The optimal number of clusters is K = 3, because after that inertia reduction is very small and clusters become less meaningful
